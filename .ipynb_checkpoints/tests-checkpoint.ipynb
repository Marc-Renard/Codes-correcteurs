```python
K.<g> = GF(8)
R.<X> = PolynomialRing(K)
A = X^3+ g*X^2+(1+g)*X+g
print(A.coefficients(sparse=False))
print(A.list())
```

```python
k=5
p=3
positions = Subsets(k, p)
Values = []
for lmt in K:
    Values.append(lmt)
Values.remove(0)
print(Values)
Index = (IntegerModRing(len(Values)))^p


for pos in positions: #emplacement des valeurs non nulles du vecteur x
    x = vector(K, [0]*k)
    for index in Index: # vecteur de (Z/(|K|-1)Z ), permet de parcourir toutes vecteurs de p dans K sans aucune coordonnée nulle
        for a in range(p):
            x[pos[a - 1]] = Values[ index[a] ]
        print("x : ",x)
```

```python
H = []
for v in K^2:
    test = 1
    for x in K:
        if x != 0:
            if x*v in H:
                test = 0
                break
    if test == 1:
        if((v[0]!=0) or (v[1]!=0)):
            H.append(v)
for i in range(len(H)):
    H[i] = H[i]*(g^2+1)
print(H)
```

```python
for I in Subsets(9,2):
    h = Matrix(K, 2)
    h[0,0]=H[I[0]-1][0]
    h[0,1]=H[I[0]-1][1]
    h[1,0]=H[I[1]-1][0]
    h[1,1]=H[I[1]-1][1]
    print(h.determinant()!=0)
    
```
